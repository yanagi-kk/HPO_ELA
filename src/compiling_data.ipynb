```python
import os
import pandas as pd
import numpy as np
```

```python
config = {
    "name": "all", # dataset, algo, param, metrics
    "select_No": ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10"], # 01~
}
output_path = f"../exp_results/for_idea/{config['name']}.csv"
```

```python
out_df = pd.DataFrame()
id_list = []
dataset_list = []
algo_list = []
param_list = []
metrics_list = []

for root, _, files in os.walk(top="..\\exp_results"):
    if root == "..\\exp_results":
        continue

    if root.split("\\")[-1] == "for_idea":
        continue

    # if root.split("\\")[-1].split("_")[0] not in config["select_No"]:
    #     continue
    
    id_list.append(root.split("\\")[-1].split("_")[0])
    dataset_list.append(root.split("\\")[-1].split("_")[1].split("×")[0])
    algo_list.append(root.split("\\")[-1].split("_")[1].split("×")[1])
    param_list.append(root.split("\\")[-1].split("_")[1].split("×")[2])
    metrics_list.append(root.split("\\")[-1].split("_")[1].split("×")[3])

    file = [file for file in files if file == "ela_features.csv"][0]
    filePath = os.path.join(root, file)
    df = pd.read_csv(filePath)
    out_df = pd.concat([out_df, df])

out_df.reset_index(drop=True, inplace=True)
```

```python
df_meta = pd.DataFrame({"id": id_list,
              "dataset": dataset_list, 
              "algorithm": algo_list, 
              "params_num": param_list, 
              "metrics": metrics_list })
```

```python
df = pd.concat([df_meta, out_df], axis=1)
df
```

```python
df.to_csv(output_path, index=False)
```
